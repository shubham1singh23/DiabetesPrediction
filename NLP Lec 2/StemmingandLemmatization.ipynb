```python
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer

data=pd.read_csv("Code_4.csv")
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer

ps=PorterStemmer()
def clean(text):
    text=text.lower()
    text=word_tokenize(text)
    text=[ps.stem(t) for t in text if t not in stopwords.words("english") ]
    text=",".join(text)
    return text

data["clean_text"]=data["review"].apply(clean)
print(data["review"])
print(data["clean_text"])
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

wnl=WordNetLemmatizer()
def clean(text):
    text=text.lower()
    text=word_tokenize(text)
    text=[wnl.lemmatize(t) for t in text if t not in stopwords.words("english") ]
    text=",".join(text)
    return text

data["clean_text"]=data["review"].apply(clean)
print(data["review"])
print(data["clean_text"])

from sklearn.feature_extraction.text import CountVectorizer
cv=CountVectorizer()
vector=cv.fit_transform(data["clean_text"])

features=pd.DataFrame(vector.toarray(),columns=cv.get_feature_names_out())
print(features)
```
