```python
import  pandas as pd
import re

from sklearn.metrics.pairwise import cosine_similarity

data=pd.read_csv("movies.csv")

def clean_function(text):
    text = text.lower()
    text = text.replace("(", "").replace(")", "")
    text = re.sub(r'\d+', '', text)
    return text.strip()

data["clean_title"]=data["title"].apply(clean_function)
print(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

title_cv=CountVectorizer()
title_vector=title_cv.fit_transform(data["clean_title"])

user_title=input("enter Movie")
clean_title=clean_function(user_title)
user_vector=title_cv.transform([clean_title])
user_frame=pd.DataFrame(user_vector)
user_frame.to_csv("Book1.csv")


cs_title=cosine_similarity(user_vector,title_vector)
movie_index=cs_title.argmax()
print(movie_index)

chosen_movie=data.iloc[movie_index]
print(chosen_movie)
print(chosen_movie["genres"])

data["clean_genres"]=data
```
