```python
import pandas as pd

data=pd.read_csv("Code_5.csv")
print(data)
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia=SentimentIntensityAnalyzer()

def getLabel(text):
    ps=sia.polarity_scores(text)
    score=ps["compound"]
    if score >= 0.05:
        return "Positive"
    elif score <=-0.05:
        return "Negative"
    else: return "Neutral"

    data["label"]=data["review"].apply(getLabel)
    print(data)
    print("Hello")
```
