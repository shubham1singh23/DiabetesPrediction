```python
from nltk.chat import Chat

conv=[
    ["hello|hi|hey",["welcome to the club ","aapka swagat hai","Hi there"]],
    ["admissions|admission|join|apply",["Starting from June","June se starting","May ke baad starting"]],
    ["fees|fee|paisa|money|amount",["Contact Office","Affordable","Scholarship avail"]]
]

chat=Chat(conv)
print("Welcome to our college")

while True:
    ques=input("Enter Sawal")
    if ques=="quit":
        print("Have a nice day")
        break
    elif chat.respond(ques) is not None:
        ans=chat.respond(ques)
        print(ans)
    else:
        print("Sorry , No response")
```
