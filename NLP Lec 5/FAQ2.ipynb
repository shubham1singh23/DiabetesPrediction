```python
import pandas as pd

data=pd.read_csv("faq.csv")
print(data)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
vector_cv=cv.fit_transform(data["qts"])
print(vector_cv)
```

```python
from sklearn.metrics.pairwise import cosine_similarity

while True:
    ques=input("Enter question")

    if ques.lower() =="quit":
        break

    user_vector=cv.transform([ques])
    cs=cosine_similarity(vector_cv,user_vector)

    cs=cs.flatten()
    print(cs)
    index=cs.argmax()
    row=data.iloc[index]
    print(row["ans"])

```
