```python
import pandas as pd
data=pd.read_csv("food.csv")
print(data)
```

```python
def clean_func(text):
    text=text.lower()
    text=text.strip()
    return text

data["clean_food_name"]=data["food_name"].apply(clean_func)
print(data["clean_food_name"])
```

```python
import spacy

nlp=spacy.load("en_core_web_lg")

data["vector_food_name"]=data["clean_food_name"].apply(nlp)

print(data["vector_food_name"])
```

```python
user_food=input("Enter the food name")
clean_user_food=clean_func(user_food)
vector_user_food=nlp(clean_user_food)

score=[]
for row in data["vector_food_name"]:
    s_score=row.similarity(vector_user_food)
    score.append(s_score)

data["s_score"]=score
max_index=data["s_score"].argmax()
row=data.iloc[max_index]
print(row)
```

```python
data["clean_matchers"]=data["matchers"].str.replace("|"," ").str.lower()
data["vector_matchers"]=data["clean_matchers"].apply(nlp)

user_matchers=row["matchers"]
clean_user_matchers=user_matchers.replace("|"," ").lower()
vector_user_matchers=nlp(clean_user_matchers)

score2=[]
for row in data["vector_matchers"]:
    s_score2=row.similarity(vector_user_matchers)
    score2.append(s_score2)

data["score2"]=score2
sorted_score=data["score2"].argsort()[::-1]
recommended_index=sorted_score[1:6]

recommended_food=data.iloc[recommended_index]
print(recommended_food["food_name"])
```
