```python

import pandas as pd


data=pd.read_csv("Code7.csv")
print(data)
```

```python
from nltk import word_tokenize
from nltk.corpus import stopwords
from string import punctuation

def clean(text):
    text=text.lower()
    text=word_tokenize(text)
    text=[x for x in text if x not in punctuation]
    text=[x for x in text if x not in stopwords.words("english")]
    text=",".join(text)
    return text
data["cleaned_review"]=data["review"].apply(clean)
print(data["cleaned_review"])
```

```python
from sklearn.feature_extraction.text import CountVectorizer

cv=CountVectorizer()
vector=cv.fit_transform(data["cleaned_review"],)
print(vector)
```

```python
features=pd.DataFrame(vector.toarray())
print(features)
```

```python
features=pd.DataFrame(vector.toarray(),columns=cv.get_feature_names_out())
print(features)
```
