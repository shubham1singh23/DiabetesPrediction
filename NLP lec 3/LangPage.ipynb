```python
from pickle import load

from matplotlib.pyplot import title

f=open("Countvector.pkl","rb")
cv=load(f)
f.close()

f=open("LangModel.pkl","rb")
model=load(f)
f.close()
```

```python
import re
import fastapi as FastAPI

def clean_function(text):
    text = re.sub(r'<[^>]+>', ' ', text)             # Remove HTML tags
    text = re.sub(r'http\S+|www\S+|@\S+', ' ', text) # Remove URLs and emails
    text = re.sub(r'\d+', ' ', text)                 # Remove numbers
    text = re.sub(r'\s+', ' ', text).strip()         # Normalize spaces
    return text

app=FastAPI(title="Language Detector")

@app.post('/detect')
def detect(text:str):
    clean_text=clean_function(text)
    vector_text=cv.transform([clean_text])
    result=model.predict(vector_text)
    print(result)
    return {"output":result[0]}




```
