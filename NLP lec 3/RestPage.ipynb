```python
from pickle import load


f=open("Countvector3.pkl","rb")
cv=load(f)
f.close()

f=open("RestModel.pkl","rb")
model=load(f)
f.close()
```

```python
import pandas as pd

data=pd.read_csv("test.tsv",sep="\t")
review=data["Review"]
vector_text=cv.transform(review)

result=model.predict(vector_text)
data["Prediction"]=result
print(data)
```
