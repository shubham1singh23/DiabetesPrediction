```python
import pandas as pd

data=pd.read_csv("Code_4.csv")
```

```python
from nltk.sentiment.vader import SentimentIntensityAnalyzer

sia=SentimentIntensityAnalyzer()

def getSentiment(text):
    ps=sia.polarity_scores(text)
    if ps["compound"]<=0.05:
        return "Positive"
    elif ps["compound"]<=-0.05 :
        return "Negative"
    else:
        return "Neutral"

data["Sentiment"]=data["review"].apply(getSentiment)
print(data)
```
